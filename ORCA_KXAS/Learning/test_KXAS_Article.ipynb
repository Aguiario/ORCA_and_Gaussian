```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_Cu_I_eV = np.loadtxt(os.path.join("KXAS_Article", "Cu_I", "KXAS_Article_Cu_I.out.absq.stk"))
x_Cu_I_eV, y_Cu_I_eV = file_Cu_I_eV[:, 0], file_Cu_I_eV[:, 1]
file_Cu_II_H2O_eV = np.loadtxt(os.path.join("KXAS_Article", "Cu_II_H2O", "KXAS_Article_Cu_II_H2O.out.absq.stk"))
x_Cu_II_H2O_eV, y_Cu_II_H2O_eV = file_Cu_II_H2O_eV[:, 0], file_Cu_II_H2O_eV[:, 1]

bar_width = 0.05

plt.figure(figsize=(10, 6))
plt.bar(x_Cu_I_eV, y_Cu_I_eV, width=bar_width, label="Cu_I", alpha=0.6)
plt.bar(x_Cu_II_H2O_eV, y_Cu_II_H2O_eV, width=bar_width, label="Cu_II_H2O", alpha=0.6)
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (a.u.)")
plt.legend()
plt.show()
```
