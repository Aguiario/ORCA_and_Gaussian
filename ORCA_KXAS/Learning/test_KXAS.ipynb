```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_Cu_I_icm = np.loadtxt(os.path.join("KXAS", "Cu_I", "KXAS_Cu_I_APP_E2_opt.out.abs.dat"))
x_Cu_I_icm, y_Cu_I_icm = file_Cu_I_icm[:, 0], file_Cu_I_icm[:, 1]
file_Cu_II_H2O_icm = np.loadtxt(os.path.join("KXAS", "Cu_II_H2O", "KXAS_Cu_II_H2O_APP_E2_opt.out.abs.dat"))
x_Cu_II_H2O_icm, y_Cu_II_H2O_icm = file_Cu_II_H2O_icm[:, 0], file_Cu_II_H2O_icm[:, 1]

bar_width = 200

plt.figure(figsize=(10, 6))
plt.bar(x_Cu_I_icm, y_Cu_I_icm, width=bar_width, label="Cu_I", alpha=0.6)
plt.bar(x_Cu_II_H2O_icm, y_Cu_II_H2O_icm, width=bar_width, label="Cu_II_H2O", alpha=0.6)
plt.xlabel("Wavenumber (cm^-1)")
plt.ylabel("Intensity (a.u.)")
plt.legend()
plt.show()
```

```python
file_Cu_I_eV = np.loadtxt(os.path.join("KXAS", "Cu_I", "KXAS_Cu_I_APP_E2_opt.out.absq.stk"))
x_Cu_I_eV, y_Cu_I_eV = file_Cu_I_eV[:, 0], file_Cu_I_eV[:, 1]
file_Cu_II_H2O_eV = np.loadtxt(os.path.join("KXAS", "Cu_II_H2O", "KXAS_Cu_II_H2O_APP_E2_opt.out.absq.stk"))
x_Cu_II_H2O_eV, y_Cu_II_H2O_eV = file_Cu_II_H2O_eV[:, 0], file_Cu_II_H2O_eV[:, 1]

bar_width = 0.03

plt.figure(figsize=(10, 6))
plt.bar(x_Cu_I_eV, y_Cu_I_eV, width=bar_width, label="Cu_I", alpha=0.6)
plt.bar(x_Cu_II_H2O_eV, y_Cu_II_H2O_eV, width=bar_width, label="Cu_II_H2O", alpha=0.6)
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (a.u.)")
plt.legend()
plt.show()
```
