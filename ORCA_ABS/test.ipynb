```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_ref_Cu_I = os.path.join("data", "Cu_I_xbpy2.tsv")
curve_ref_Cu_I, bar_ref_Cu_I = utl.plot(file_ref_Cu_I, plot=False,sigma=0.25)


plt.figure(figsize=(10, 6))
plt.plot(curve_ref_Cu_I[0], curve_ref_Cu_I[1], label="$Cu_I$", color='blue', linewidth=2)
plt.bar(bar_ref_Cu_I[0], bar_ref_Cu_I[1], width=0.1, color='skyblue', label="Transitions $Cu_I$")
plt.xlim(8978,8986)
plt.ylim(0,7)
plt.grid(True)
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")


# Mostrar el gráfico
plt.show()
```

```python
file_ref_Cu_II = os.path.join("data", "Cu_II_xbpy2.tsv")
curve_ref_Cu_II, bar_ref_Cu_II = utl.plot(file_ref_Cu_II, plot=False,sigma=1)
file_ref_Cu_II_ClO4 = os.path.join("data", "Cu_II_xbpy2_ClO4-2.tsv")
curve_ref_Cu_II_ClO4, bar_ref_Cu_II_ClO4 = utl.plot(file_ref_Cu_II_ClO4, plot=False,sigma=1)



plt.figure(figsize=(10, 6))
plt.plot(curve_ref_Cu_II[0]-2, curve_ref_Cu_II[1], label="$[Cu(xbpy)_2]^{+2}$", color='purple', linewidth=2)
plt.plot(curve_ref_Cu_II_ClO4[0]-2, curve_ref_Cu_II_ClO4[1], label="$[Cu(xbpy)_2]^{+2} - (ClO_4)_2$", color='green', linewidth=2)

plt.xlim(8975,9010)
plt.legend()
plt.grid(True)
#plt.ylim(0,7)
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")


# Mostrar el gráfico
plt.show()
```
