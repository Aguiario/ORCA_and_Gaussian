```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_path_Cu_I = os.path.join("data", "Cu_I_nr40.tsv")
curve_Cu_I, bar_Cu_I = utl.plot(file_path_Cu_I)
```

```python
file_path_Cu_II = os.path.join("data", "Cu_II_nr40.tsv")
curve_Cu_II, bar_Cu_II = utl.plot(file_path_Cu_II)
```

```python
selected_Cu_II = [0,1, 2, 3, 4, 6, 8, 10, 13,15,17, 19, 22, 25, 30, 35, 39]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(curve_Cu_II[0], curve_Cu_II[1], label="Gaussian Lineshape", color='blue', linewidth=2)
plt.bar(bar_Cu_II[0][selected_Cu_II], bar_Cu_II[1][selected_Cu_II], width=0.01, color='orange', label="Datos originales (sticks)")
# Configuracin del grfico
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (a.u.)")
plt.grid(True)

# Mostrar el grfico
plt.show()
```
