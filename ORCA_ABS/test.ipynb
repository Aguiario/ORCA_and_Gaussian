```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_ref_Cu_I = os.path.join("data", "Cu_I_xbpy2.tsv")
curve_ref_Cu_I, bar_ref_Cu_I = utl.plot(file_ref_Cu_I, plot=False,sigma=0.25)


plt.figure(figsize=(10, 6))
plt.plot(curve_ref_Cu_I[0], curve_ref_Cu_I[1], label="$Cu_I$", color='blue', linewidth=2)
plt.bar(bar_ref_Cu_I[0], bar_ref_Cu_I[1], width=0.01, color='skyblue', label="Transitions $Cu_I$")
plt.xlim(8978,8986)
plt.ylim(0,7)
plt.grid(True)
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")


# Mostrar el gráfico
plt.show()
```

```python
file_ref_Cu_II = os.path.join("data", "Cu_II_xbpy2_ClO4-2.tsv")
curve_ref_Cu_II, bar_ref_Cu_II = utl.plot(file_ref_Cu_I, plot=False,sigma=0.25)


plt.figure(figsize=(10, 6))
plt.plot(curve_ref_Cu_II[0], curve_ref_Cu_II[1], label="$Cu_II$", color='red', linewidth=2)
plt.bar(bar_ref_Cu_II[0], bar_ref_Cu_II[1], width=0.01, color='orange', label="Transitions $Cu_II$")
plt.xlim(8978,8986)
plt.ylim(0,7)
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")


# Mostrar el gráfico
plt.show()
```
