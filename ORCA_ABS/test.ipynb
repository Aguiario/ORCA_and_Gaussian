```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
selected = [0,1, 2, 3, 4, 6, 8, 10, 13,15,17, 19, 22, 25, 27, 30, 33, 35, 37]

file_path_Cu_I = os.path.join("data", "Cu_I_UVVIS_nr40.tsv")
curve_Cu_I, bar_Cu_I = utl.plot(file_path_Cu_I, plot=False)

file_path_Cu_II = os.path.join("data", "Cu_II_UVVIS_nr40.tsv")
curve_Cu_II, bar_Cu_II = utl.plot(file_path_Cu_II,plot=False)

plt.figure(figsize=(10, 6))
plt.plot(curve_Cu_I[0], curve_Cu_I[1], label="$Cu_I$", color='blue', linewidth=2)
plt.bar(bar_Cu_I[0][selected], bar_Cu_I[1][selected], width=0.01, color='skyblue', label="Transitions $Cu_I$")
plt.plot(curve_Cu_II[0], curve_Cu_II[1], label="$Cu_{II}$", color='red', linewidth=2)
plt.bar(bar_Cu_II[0][selected], bar_Cu_II[1][selected], width=0.01, color='orange', label="Transitions $Cu_{II}$")
# Configuración del gráfico
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
reference_Cu_I = os.path.join("data", "test_11_XRAY.tsv")
Cu_I_XRAY = os.path.join("data", "Cu_I_XRAY.tsv")
curve_reference_Cu_I, bar_reference_Cu_I = utl.plot(reference_Cu_I,plot=False, sigma = 0.8)
curve_Cu_I_XRAY, bar_Cu_I_XRAY = utl.plot(Cu_I_XRAY, plot=False, sigma=0.8)

plt.figure(figsize=(10, 6))
plt.plot(curve_reference_Cu_I[0], curve_reference_Cu_I[1], label="ref $Cu_I$", color='blue', linewidth=2)
plt.bar(bar_reference_Cu_I[0], bar_reference_Cu_I[1], width=0.1, color='skyblue', label="Transitions ref $Cu_I$")

plt.plot(curve_Cu_I_XRAY[0], curve_Cu_I_XRAY[1], label="$Cu_I$", color='green', linewidth=2)
plt.bar(bar_Cu_I_XRAY[0], bar_Cu_I_XRAY[1], width=0.1, color='green', label="Transitions $Cu_I$")
# Configuración del gráfico
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
Cu_I_nm = os.path.join("data", "Cu_I_UVVis_nm_nr40.tsv")
Cu_II_nm = os.path.join("data", "Cu_II_UVVis_nm_nr40.tsv")
curve_Cu_I_nm = np.loadtxt(Cu_I_nm, skiprows=1)
curve_Cu_II_nm = np.loadtxt(Cu_II_nm, skiprows=1)

plt.figure(figsize=(10, 6))
plt.plot(curve_Cu_I_nm[:,0], curve_Cu_I_nm[:,1], label="$Cu_{I}$", color='blue', linewidth=2)
plt.plot(curve_Cu_II_nm[:,0], curve_Cu_II_nm[:,1], label="$Cu_{II}$", color='red', linewidth=2)
# Configuración del gráfico
plt.xlabel("wavelength (nm)")
plt.ylabel("Absorbance (a.u.)")
plt.legend()

# Mostrar el gráfico
plt.show()
```
