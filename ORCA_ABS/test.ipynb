```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Lista todos los archivos en la carpeta
for file_name in os.listdir("data"):
    # Verifica si el archivo tiene la extensión .tsv
    if file_name.endswith(".tsv"):
        # Construye la ruta completa del archivo
        file_path = os.path.join("data", file_name)
        
        # Usa el nombre del archivo (sin extensión) como el argumento `name`
        name = os.path.splitext(file_name)[0]
        
        # Llama a la función plot
        curve, bar = utl.plot(file_path, name=name)
```

```python
file_path_Cu_II = os.path.join("data", "Cu_II_nr40.tsv")
curve_Cu_II, bar_Cu_II = utl.plot(file_path_Cu_II)
```

```python
selected_Cu_II = [0,1, 2, 3, 4, 6, 8, 10, 13,15,17, 19, 22, 25, 30, 35, 39]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(curve_Cu_II[0], curve_Cu_II[1], label="Gaussian Lineshape", color='blue', linewidth=2)
plt.bar(bar_Cu_II[0][selected_Cu_II], bar_Cu_II[1][selected_Cu_II], width=0.01, color='orange', label="Datos originales (sticks)")
# Configuración del gráfico
plt.xlabel("Energy (eV)")
plt.ylabel("Intensity (a.u.)")
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
