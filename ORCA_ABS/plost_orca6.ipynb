```python
import ORCA_plot as utl
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
sigma = 0.25
file_Cu_I = os.path.join("data", "Cu_I.tsv")
curve_Cu_I, bar_Cu_I = utl.plot(file_Cu_I,plot=False,sigma=sigma)

file_Cu_I_T2_H = os.path.join("data_orca6", "Cu_I_T2_H.tsv")
curve_Cu_I_T2_H, bar_Cu_I_T2_H = utl.plot(file_Cu_I_T2_H,plot=False,sigma=sigma)

file_Cu_I_T2U_H = os.path.join("data_orca6", "Cu_I_T2_H.tsv")
curve_Cu_I_T2U_H, bar_Cu_I_T2U_H = utl.plot(file_Cu_I_T2U_H,plot=False,sigma=sigma)



plt.figure(figsize=(10, 6))
plt.plot(curve_Cu_I[0]-11.145, curve_Cu_I[1], label="$Cu^{I}$", color='red', linewidth=2)
plt.plot(curve_Cu_I_T2U_H[0]-11.4, curve_Cu_I_T2U_H[1], label="$Cu^{I} - T2$", color='blue', linewidth=2)
plt.plot(curve_Cu_I_T2_H[0]-11.4, curve_Cu_I_T2_H[1], label="$Cu^{I} - T2U$", color='green', linewidth=2)


plt.legend()
plt.grid(True)
#plt.ylim(0,7)
plt.xlabel("Energy (eV)")
plt.ylabel("Absorbance (a.u.)")


# Mostrar el gráfico
plt.xlim(8979,8983.5)
plt.show()
```
