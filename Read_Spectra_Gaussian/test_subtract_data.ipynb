```python
import spectra
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the paths to the files containing IR data
file_paths_B_Raman = ['B\\B_Raman_Spin3_180.txt',
                      'B\\B_Raman_spin3_sixexchangable.txt']

# Initialize a list to store the intensity values (y)
intensities = []

# Loop through each file and extract the data
for file in file_paths_B_Raman:
    x, y = spectra.subtract_data(file)  # Get the Raman shift (x) and intensity (y) data
    intensities.append(y)  # Append the intensity data to the list

# Convert the list of intensities to a numpy array for easier averaging
intensities = np.array(intensities)

# Calculate the average intensity across all files
average_intensity = np.mean(intensities, axis=0)

# Plot the average intensity in red
plt.plot(x, average_intensity, 'g-', linewidth=2)

# Add labels and a title to the plot
plt.xlabel('Raman Shift (cm)') 
plt.ylabel('Normalized Intensity')
plt.title('Raman Spectra')  

# Display the plot
plt.show()
```

```python
# Define the paths to the files containing IR data
file_paths_B = ['B\\B_IR_spin3_180.txt',
                'B\\B_IR_spin3_singlepoint.txt',
                'B\\B_IR_spin3_sixexchangable.txt',
                'B\\B_IR_spin3.txt']

# Initialize a list to store the intensity values (y)
intensities = []

# Loop through each file and extract the data
for file in file_paths_B:
    x, y = spectra.subtract_data(file)  # Get the Raman shift (x) and intensity (y) data
    intensities.append(y)  # Append the intensity data to the list

# Convert the list of intensities to a numpy array for easier averaging
intensities = np.array(intensities)

# Calculate the average intensity across all files
average_intensity = np.mean(intensities, axis=0)

# Plot the average intensity in red
plt.plot(x, average_intensity, 'r-', linewidth=2)
plt.axvspan(450, 750, color='orange', alpha=0.5)
plt.text((450 + 750) / 2, 0.8, 'Mn-O-Mn', color='orange', ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
plt.axvspan(950, 1050, color='skyblue', alpha=0.5)
plt.text((950 + 1050) / 2, 0.8, 'OEC', color='blue', ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

plt.text(500, 0.31, '-O', color='black', ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
plt.text(700, 0.32, 'Mn1-O5-Mn3', color='black', ha='center', fontsize=10, bbox=dict(facecolor='white', alpha=0.7))

# Add labels and a title to the plot
plt.xlabel('Wavenumber (cm)') 
plt.ylabel('Normalized Intensity')
plt.title('Calculated IR Spectra -O Model')  

# Display the plot
plt.show()
```
