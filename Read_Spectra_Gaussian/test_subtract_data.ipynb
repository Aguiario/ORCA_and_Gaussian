```python
import spectra
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the paths to the files containing IR data
file_paths_B_Raman = ['B\\B_Raman_Spin3_180.txt',
                      'B\\B_Raman_spin3_sixexchangable.txt']

# Initialize a list to store the intensity values (y)
intensities = []

# Loop through each file and extract the data
for file in file_paths_B_Raman:
    x, y = spectra.subtract_data(file)  # Get the Raman shift (x) and intensity (y) data
    intensities.append(y)  # Append the intensity data to the list

# Convert the list of intensities to a numpy array for easier averaging
intensities = np.array(intensities)

# Calculate the average intensity across all files
average_intensity = np.mean(intensities, axis=0)

# Plot the average intensity in red
plt.plot(x, average_intensity, 'g-', linewidth=2)

# Add labels and a title to the plot
plt.xlabel('Raman Shift (cm)') 
plt.ylabel('Normalized Intensity')
plt.title('Raman Spectra')  

# Display the plot
plt.show()
```
