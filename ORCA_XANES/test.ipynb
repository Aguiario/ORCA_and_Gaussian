```python
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = np.loadtxt("dat.dat")
x = data[:, 0]  # Wavelength/Energy (eV)
lambda_nm = 1240 / x  # Convert energy (eV) to wavelength (nm)

y = data[:, 1]  # Absorption intensity

# Plot the spectrum
plt.plot(lambda_nm, y, label="Absorption Spectrum")
#plt.xlabel("Energy (eV)")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Intensity (a.u.)")
plt.title("TD-DFT Absorption Spectrum")
plt.legend()
plt.show()
```
